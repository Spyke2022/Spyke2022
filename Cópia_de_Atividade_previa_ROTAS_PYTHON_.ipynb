```python
import pandas as pd
import numpy as np
df = pd.read_csv('/content/rotas.csv')
df.loc[1] = ['class', 'meta', 'feature', 'feature','feature', 'feature', 'feature', 'feature', 'feature']
df.head(20)
```

```python
df.tail()
```

```python
df['ROTAS'] = df['ROTAS'].str.replace('R', '')
df.head()
```

```python
df = df.drop('POSTOPOLICIAL', axis=1)
```

```python
df["DIRIGIBILIDADE"] = pd.to_numeric(df["DIRIGIBILIDADE"], errors='coerce')
df["ROTAS"] = pd.to_numeric(df["ROTAS"], errors='coerce')
df["DISTANCIA"] = pd.to_numeric(df["DISTANCIA"], errors='coerce')
df["FALHAS"] = pd.to_numeric(df["FALHAS"], errors='coerce')
df["LOMBADAS"] = pd.to_numeric(df["LOMBADAS"], errors='coerce')
df["ILUMINACAO"] = pd.to_numeric(df["ILUMINACAO"], errors='coerce')
df["SINALIZACAO"] = pd.to_numeric(df["SINALIZACAO"], errors='coerce')
df["PEDAGIOS"] = pd.to_numeric(df["PEDAGIOS"], errors='coerce')
print(df.info())
df.isna().sum()
```

```python
df.dropna(inplace = True)
print(df.info())
df.isna().sum()
```

```python
df_analise = df.iloc[1:]
```

```python
df_analise.sort_values(by=['FALHAS'], ascending=True, inplace=False).head(10)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
# Criar e treinar o modelo K-Means
kmeans = KMeans(n_clusters=4)
kmeans.fit(df_analise)

# Obter rótulos dos clusters e coordenadas dos centróides
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
df_analise["CLUSTER"] = kmeans.labels_
df_analise.head()
```

```python
df_analise["CLUSTER"].value_counts()
```

```python
correlation = pd.DataFrame(df_analise.corr())
correlation
```

```python
from matplotlib import pyplot as plt
correlation['ROTAS'].plot(kind='hist', bins=20, title='ROTAS')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
from matplotlib import pyplot as plt
correlation['ROTAS'].plot(kind='hist', bins=20, title='ILUMINACAO')
plt.gca().spines[['top', 'right',]].set_visible(False)
```
